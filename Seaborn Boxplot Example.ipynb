# Boxplot Example with Seaborn

The examples below are directly lifted from the [Seaborn documentations on boxplots](https://seaborn.pydata.org/generated/seaborn.boxplot.html).

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Get the titanic dataset:

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

Let's create the simplest boxplot:

```python
sns.boxplot(x=titanic["age"])
plt.grid()
```

Group ages by class, so that we can compare them:

```python
sns.boxplot(data=titanic, x="age", y="class")
plt.grid()
```

Now, for each class, show who was alive after the Titanic sank:

```python
sns.boxplot(data=titanic, x="class", y="age", hue="alive")
```

Add space between bars and remove fill color:

```python
sns.boxplot(data=titanic, x="class", y="age", hue="alive", fill=False, gap=.1)
```
